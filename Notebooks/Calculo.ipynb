```python
from sympy import *
init_printing()
x = Symbol('x')
datos=[
    (x**2,0,3),
    (x**2,-3,3),
    (4*x**3,0,2),
    (4*x**3,-2,2),
    (5*x**4,0,1),
    (5*x**4,-1,1)
]
for expr,a,b in datos:
    integral=Integral(expr,(x,a,b))
    resultado=integral.doit()
    display(Eq(integral,resultado))
```
